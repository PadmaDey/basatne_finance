#### difference calculation

1. I need to select a base ['Country'] for ['Price (USD)']
2. Then group by based on ['Model'] and ['Size(GB)']
3. Then from the ['Price (USD)'] of all the ['Country'], I need to deduct the base ['Country']'s ['Price (USD)']  

  (if base_price == 0.0:
  base_price = the next price)
4. Then calculate the percentage : (difference/ base price)*100

```python
df_copy = df.copy()
df_copy.sample(2)
```

```python
base_country = 'United States'

grouped = df.groupby(['Model', 'Size(GB)', 'Color'])

df['difference'] = 0.0
df['percentage_diff'] = 0.0

for (model, size, color), group in grouped:
    base_price = group.loc[group['Country'] == base_country, 'Price (USD)']
    if not base_price.empty:
        base_price = base_price.iloc[0]
        df.loc[group.index, 'difference'] = group['Price (USD)'] - base_price
        df.loc[group.index, 'percentage_diff'] = (
            df.loc[group.index, 'difference'] / base_price * 100
        )

df.sample(10)
```
